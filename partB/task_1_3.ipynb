## Baseline Method: Standard Gaussian Mixture Model with EM

**Description**

The baseline method is the standard Gaussian Mixture Model (GMM) trained using the Expectation Maximization (EM) algorithm. In this approach, the dataset is assumed to be generated from a mixture of Gaussian distributions. Each Gaussian component represents a cluster.

The EM algorithm iteratively performs two steps:

1. **Expectation Step:** Compute the probability that each data point belongs to each cluster.
2. **Maximization Step:** Update the parameters of the Gaussian components (means, covariances, and mixture weights) based on these probabilities.

This process continues until the log-likelihood converges.

## Limitation of the Baseline Method

**Limitation**

The standard EM algorithm relies entirely on the statistical structure of the data and ignores any external knowledge about relationships between data points.

Because of this, the algorithm may produce clustering results that maximize likelihood but do not reflect the true underlying grouping of the data.

For example, when clusters overlap or when the Gaussian assumption leads to ambiguous boundaries, the EM algorithm may incorrectly partition the data.

---

## How the Paper Improves the Baseline

**Improvement**

The paper improves the baseline by incorporating equivalence constraints between data points into the EM framework.

These constraints specify whether pairs of data points originate from the same source or from different sources.

Positive constraints are used to construct chunklets, which group together points that must share the same cluster assignment. The EM algorithm is then modified so that cluster probabilities are computed for chunklets instead of individual points.

This modification allows the algorithm to incorporate side information directly into the likelihood function, guiding the clustering process toward solutions that respect the provided constraints.

---

## When the Proposed Method May Not Outperform the Baseline

**Failure Scenario**

The constrained EM algorithm may perform worse than the baseline if the equivalence constraints are incorrect or noisy.

If points from different true clusters are mistakenly grouped into the same chunklet, the algorithm will force them to share the same cluster assignment. This can distort the estimated Gaussian parameters and lead to poorer clustering results than the unconstrained EM algorithm.

---

## Summary

The baseline method is the standard EM algorithm for Gaussian mixture models, which performs clustering based solely on the statistical distribution of the data. The proposed method improves upon this baseline by incorporating equivalence constraints that guide cluster assignments. However, the effectiveness of this approach depends on the correctness of the constraints, and incorrect constraints can degrade performance.
